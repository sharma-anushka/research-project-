```python
import pandas as pd
```

```python
data = pd.read_csv('data_final.csv.csv')
data
```

```python
data = data.drop('SERVICES', axis='columns')
data
```

```python
data = data.drop("REASON", axis='columns')
data
```

```python
# Assuming 'data' is a Pandas DataFrame and 'LOS' is a column
data['LOS'] = data['LOS'].apply(lambda x: 0 if x <= 20 else 1)
print(data['LOS'].unique())
```

```python
data
```

```python
data = data[data['GENDER'] != -9]
```

```python
data
```

```python
contains_neg9 = (data['LOS'] == 1).any()
print(contains_neg9)
```

```python
data = data[data['NOPRIOR'] != -9]
data = data[data['PSYPROB'] != -9]
data = data[data['PREG'] != -9]
data = data[data['AGE'] != -9]
data = data[data['DSMCRIT'] != -9]
data = data[data['SUB1'] != -9]
data = data[data['SUB2'] != -9]
data = data[data['SUB3'] != -9]
data = data[data['FREQ1'] != -9]
data = data[data['FREQ2'] != -9]
data = data[data['FREQ3'] != -9]
data = data[data['ALCDRUG'] != -9]
```

```python
data
```

```python
y = data['LOS']
y
```

```python
inputs = data[['GENDER', 'NOPRIOR', 'PSYPROB', 'PREG', 'AGE', 'DSMCRIT', 'SUB1', 'SUB3', 'SUB2', 'FREQ1','FREQ3', 'FREQ2','ALCDRUG']]
inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, y, train_size = 0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
# decision trees

from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
```
